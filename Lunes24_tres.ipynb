solución de la ecuación:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.975, 1.05, 0.001)
y = []

# solución forma 1
def fun1(a):
    return 15*3**(a - 1) + 3**(a + 1) + 3**a

for a in x:
    y.append(fun1(a))

f2 = np.full(x.size, 27)
print(x)

# gráfica forma 1
#forma 1
plt.plot(x, y, 'r', x, f2, 'y')

#solución forma 2
f1 = 15*3**(x - 1) + 3**(x + 1) + 3**x
f2 = np.full(x.size,27)

#gráfica forma 2
fig, ax = plt.subplots()
ax.plot(x, f1, 'b', x, f2, 'm')
ax.grid('on')
plt.show()
```
