```python
import numpy as np
import matplotlib.pyplot as pl
paso = 0.01
p = np.arange(-1,1,paso)
ft = np.sin(np.sqrt(3)*p)*np.tan(p)

pl.plot(p,ft, 'r')
pl.fill_between(p,ft,color=(1,0.616,0,0.459))
pl.title("sin[3^(1/2)t]tan[t]")
pl.show()

AreaFt = np.sum(paso*ft)

import math

fx = []
area2 = 0
def fun1(x):
  return math.sin(math.sqrt(3)*x)*math.tan(x)

for h in p:
  y = fun1(h)
  fx.append(y)
  area2 += y*paso

fig, ax = pl.subplots()
ax.plot(p,fx)
ax.fill_between(p,fx,color=(1,0.616,0,0.459))

from sympy import Integral, Symbol, sqrt, sin, tan

w = Symbol('w')

integral = Integral(sin(sqrt(3)*w)*tan(w), (w, -1, 1))

print("el área bajo la curva es ",integral.evalf())
```
