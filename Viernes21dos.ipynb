```python
ls = numpy.linspace(0, 10*numpy.pi, 100)
ar = numpy.arange(0, 10*numpy.pi, 0.1)
print(ls)
print(ar)
```

```python
import matplotlib.pyplot as plt
res = numpy.sin(ls)
plt.plot(ls, res)
plt.show()
```

```python
import matplotlib.pyplot as plt
res = numpy.cos(ls)
plt.plot(ls, res)
plt.show()
```

```python
import matplotlib.pyplot as plt
res = numpy.tan(ls)
plt.plot(ls, res)
plt.show()
```

```python
import matplotlib.pyplot as plt
res = numpy.sin(ls) + numpy.tan(ls)
plt.plot(ls, res)
plt.show()
```

```python
def sumaComplejos(z1,z2):
  return(z1[0]+z2[0],z1[1]+z2[1])

def multiplicacionComplejos(z1,z2):
  return(z1[0]*z2[0]-z1[1]*z2[1],z1[0]*z2[1]+z1[1]*z2[0])
  print(sumaComplejos(z1,z2))
```
