ACTIVIDAD.
Problema Interprepas Matemáticas VI área 1
sean las funciones f(x)=sqrt(-x) y g(x)=ln(x)
graficar f ° g  ;  g ° f  ;  fg

```python
from sympy import sin, symbols, Abs, sqrt, ln, oo, Interval
from sympy.plotting import plot

def f(x):
    return sqrt(-x)
def g(x):
    return ln(x)

domain_f = Interval.open(-oo, 0)
domain_g = Interval.open(0, oo)

common_domain = domain_f.intersect(domain_g)

x = symbols('x')

plot(f(g(x)), (x, -1, 1))
plot(g(f(x)), (x, -1, 1))

if common_domain.is_empty:
    print("No se puede graficar f(x) *  g(x), porque el domino común es un conjunto vacío")
else:
    plot(f(x) * g(x), (x, common_domain))
```

ACTIVIDAD.
graficar las siguientes funciones implícitas:

*   x^2+y^2-2y=3
*   (P+5/V^2)(V - 0.03)=9.7

```python
from sympy import symbols, Eq
from sympy import plot_implicit
x, y = symbols('x,y')
print("La gráfica de la función implícita x^2 + y^2 - 2y = 3 es: ")
plot_implicit(Eq(x**2 + y**2 - 2*y - 3, 0), (x, -3, 2.5), (y, -3, 3.5));
```

```python
from sympy import symbols, Eq
from sympy import plot_implicit
P, V = symbols('P,V')
print("La gráfica de la función implícita (P+(5/V**2))*(V - 0.03) = 9.7 es: ")
eq = Eq((P+(5/V**2))*(V - 0.03), 9.7)
plot_implicit(eq, (P,-10,6),(V,-100,100));
```

ACTIVIDAD.
Grafica las siguientes funciones parametrizadas
*   x(t)=cos(t)+1/2cos(7t)+1/3sin(17t)
*   y(t)=sin(t)+1/2sin(7t)+1/3cos(17t)



```python

from sympy import symbols, cos, sin, exp, pi
from sympy import plot_parametric
t = symbols('t')
def x(t):
  return cos(t) + (1/2)*cos(7 * t) + (1/3)*sin(17 * t)
def y(t):
  return sin(t) + (1/2)*sin(7 * t) + (1/3)*cos(17 * t)

plot_parametric(x(t), y(t), (t, 0, 14*pi));
plot_parametric(x(t), y(t), (t, 0, 14*pi));
```
