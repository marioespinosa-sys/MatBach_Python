```python
from sympy import symbols, integrate, sin, exp
x = symbols('x')
integrate(sin(x), x)
```

```python
from sympy import symbols, integrate, sin, exp
x = symbols('x')
integrate(sin(x), x, (x, -1, 1))
```

```python
from sympy import symbols, integrate, sin, exp
from sympy.plotting import plot
x = symbols('x')
print(integrate(sin(x), x, (x, -1, 1)))
plot(sin(x), (x, -1, 1))
```
